```python
import pandas as pd

#Pre-processing the DIST cohort 
dist_cohort_clin = pd.read_excel('G:\diva\EP-research\Maarten Fleur\ECG features_for_manuscript\DISTANT\ECG\ECG_DICOM_all_14102022\selected_dataset_downsampled_final.xlsx')
dist_cohort_clin = dist_cohort_clin.rename(columns={'Potassium_diuretics' : 'Aldosteronremmer',
                                 'DOAC' : 'NOAC',
                                'Bblocker' : 'Betablokker',
                                 'Amiodaron' : 'Antiaritmica_soort#Amiodarone',
                                 'Digoxine' : 'Antiaritmica_soort#Digoxine',
                                'Sotalol' :'Antiaritmica_soort#Sotalol',
                                'VitK' : 'Vitamine_K'})

dist_cohort_clin['Python_ID_Index'] = dist_cohort_clin.reset_index().index
                                           
def arb(dist_cohort_clin):
    if (dist_cohort_clin['ACE']  > 0 or dist_cohort_clin['ATII_anta'] > 0 ):
        return '1'
    elif (dist_cohort_clin['ACE'] < 1 or dist_cohort_clin['ATII_anta'] < 1):
        return '0'
def Plaatjes_aggregatie(dist_cohort_clin):
    if (dist_cohort_clin['Clopidogrel_ticagrelor_prasugrel']  > 0 or dist_cohort_clin['Ascal'] > 0 ):
        return '1'
    elif (dist_cohort_clin['Clopidogrel_ticagrelor_prasugrel'] < 1 or dist_cohort_clin['Ascal'] < 1):
        return '0'  
dist_cohort_clin['ARB'] = dist_cohort_clin.apply(arb, axis = 1)
dist_cohort_clin['Plaatjes_aggregatie'] = dist_cohort_clin.apply(Plaatjes_aggregatie, axis = 1)


var = ['StudyID_x', 'Python_ID_Index', 'Age','Sex', 'AtrialArrhythmia', 'LowLVEF', 'ReducedLVEF', 'NormalLVEF', 'OHCA', 'ICM', 'NICM', 'HCM', 'DCM', 'PAS',
       'PCI', 'CABG', 'MyocardialInfarction', 'VF', 'NSVT','SustVT', 'CVA','COPD', 'DiabetesMellitus', 'BMI', 'PAF', 
       'QRS_Duration', 'Hypertension', 'CHD', 'Sodium', 'Potassium', 'Kreatinine', 'Dyslipedaemia',
          'VR', 'DR', 'CRTD', 'SICD', 'Vitamine_K', 'Antiaritmica_soort#Sotalol','ImplantationDate',
       'Antiaritmica_soort#Digoxine', 'Antiaritmica_soort#Amiodarone', 'Betablokker', 'NOAC', 'Aldosteronremmer',
       'Implantation_indication', 'Mortality', 'MtillDeath_censored', 'AppropriateTherapy','MonthsTillAppropriateTherapy']

dist_cohort_clin = dist_cohort_clin[var]

#Pre-procesing the DR cohort
dr_cohort_clin = pd.read_excel(r'G:\diva\EP-research\Maarten Fleur\ECG features_for_manuscript\DEEP RISK\CLINICAL\DR_clinical_set.xlsx')
ecg_log = pd. read_csv(r'G:\diva\EP-research\Maarten Fleur\ECG features_for_manuscript\DEEP RISK\ECG\ecgs_log.csv')

dr_cohort_clin = pd.merge(ecg_log, dr_cohort_clin, how="left", on=["PatientID"])
dr_cohort_clin['Python_ID_Index'] = dr_cohort_clin.reset_index().index
dr_cohort_clin['Python_ID_Index'] = dr_cohort_clin['Python_ID_Index'] +1799

var_bin = ['Python_ID_Index', 'Age','Sex', 'AtrialArrhythmia', 'LowLVEF', 'ReducedLVEF', 'NormalLVEF', 'OHCA', 'ICM', 
           'NICM', 'HCM', 'DCM', 'PAS', 'PCI', 'CABG', 'MyocardialInfarction', 'VF', 'NSVT','SustVT', 
           'CVA','COPD', 'DiabetesMellitus', 'BMI', 'PAF', 'QRS_Duration', 'Hypertension', 'CHD', 
           'Lab_natrium', 'Lab_kalium', 'Lab_kreatinine', 'Hypercholesterolaemie', 'Indicatie_ICD_implantatie',
           'Soort_ICD_device', 'Vitamine_K', 'Antiaritmica_soort#Sotalol', 'Date_ICD_implantatie',
           'Antiaritmica_soort#Digoxine', 'Antiaritmica_soort#Amiodarone', 'Betablokker', 'NOAC', 'Aldosteronremmer',
          'Mortality', 'Days2Death', 'AppropriateTherapy','MonthsTillAppropriateTherapy']

dr_cohort_clin = dr_cohort_clin[var_bin]


dr_cohort_clin = dr_cohort_clin.rename(columns={'Lab_natrium' : 'Sodium',
                                 'Lab_kalium' : 'Potassium',
                                 'Lab_kreatinine' : 'Kreatinine',
                                 'Hypercholesterolaemie': 'Dyslipedaemia',
                                 'Indicatie_ICD_implantatie' : 'Implantation_indication',
                                'Date_ICD_implantatie' :'ImplantationDate' })

#Allign the coding of implanted device with DIST cohort
def flag_df_vr(bin_var):
    if (bin_var['Soort_ICD_device']  == 1):
        return '1'
    elif (bin_var['Soort_ICD_device'] == 2):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 3):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 4):
        return '0'
def flag_df_dr(bin_var):
    if (bin_var['Soort_ICD_device']  == 2):
        return '1'
    elif (bin_var['Soort_ICD_device'] == 1):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 3):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 4):
        return '0' 
def flag_df_crtd(bin_var):
    if (bin_var['Soort_ICD_device']  == 3):
        return '1'
    elif (bin_var['Soort_ICD_device'] ==1):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 2):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 4):
        return '0'   
def flag_df_sicd(bin_var):
    if (bin_var['Soort_ICD_device']  == 4 ):
        return '1'
    elif (bin_var['Soort_ICD_device'] == 1):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 2):
        return '0'
    elif (bin_var['Soort_ICD_device'] == 3):
        return '0'
      
dr_cohort_clin['VR'] = dr_cohort_clin.apply(flag_df_vr, axis = 1)
dr_cohort_clin['DR'] = dr_cohort_clin.apply(flag_df_dr, axis = 1)
dr_cohort_clin['CRTD'] = dr_cohort_clin.apply(flag_df_crtd, axis = 1)
dr_cohort_clin['SICD'] = dr_cohort_clin.apply(flag_df_sicd, axis = 1)
dr_cohort_clin = dr_cohort_clin.drop('Soort_ICD_device', axis='columns')

#Convert days to months
dr_cohort_clin['MtillDeath_censored'] = (dr_cohort_clin['Days2Death'] / 31)
dr_cohort_clin = dr_cohort_clin.drop('Days2Death', axis='columns')

#Join two two cohort to one dataset
frames = [dist_cohort_clin, dr_cohort_clin]
dr_dist_cohort = pd.concat(frames, ignore_index=True)

#Creating of labels
dfObj = dr_dist_cohort
def AppropriateTherapy5Y(dfObj):
    if (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] <= 60):
        return '1'
    elif (dfObj['AppropriateTherapy'] == 0):
        return '0'
    elif (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] > 60):
        return '0'
    
def AppropriateTherapy4Y(dfObj):
    if (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] <= 48):
        return '1'
    elif (dfObj['AppropriateTherapy'] == 0):
        return '0'
    elif (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] > 36):
        return '0'

def AppropriateTherapy3Y(dfObj):
    if (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] <= 36):
        return '1'
    elif (dfObj['AppropriateTherapy'] == 0):
        return '0'
    elif (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] > 36):
        return '0'

def AppropriateTherapy2Y(dfObj):
    if (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] <= 24):
        return '1'
    elif (dfObj['AppropriateTherapy'] == 0):
        return '0'
    elif (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] > 24):
        return '0'

def AppropriateTherapy1Y(dfObj):
    if (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] <= 12):
        return '1'
    elif (dfObj['AppropriateTherapy'] == 0):
        return '0'
    elif (dfObj['AppropriateTherapy'] == 1) and (dfObj['MonthsTillAppropriateTherapy'] > 12):
        return '0'

      
dfObj['AppropriateTherapy5Y'] = dfObj.apply(AppropriateTherapy5Y, axis = 1)
dfObj['AppropriateTherapy4Y'] = dfObj.apply(AppropriateTherapy4Y, axis = 1)
dfObj['AppropriateTherapy3Y'] = dfObj.apply(AppropriateTherapy3Y, axis = 1)
dfObj['AppropriateTherapy2Y'] = dfObj.apply(AppropriateTherapy2Y, axis = 1)
dfObj['AppropriateTherapy1Y'] = dfObj.apply(AppropriateTherapy1Y, axis = 1)


dfObj['MtillDeath_censored'] = dfObj['MtillDeath_censored'].fillna(0)

def Mortality5Y(dfObj):
    if (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] <= 60):
        return '1'
    elif (dfObj['Mortality'] == 0):
        return '0'
    elif (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] > 60):
        return '0'
    
def Mortality4Y(dfObj):
    if (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] <= 48):
        return '1'
    elif (dfObj['Mortality'] == 0):
        return '0'
    elif (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] > 48):
        return '0'
    
def Mortality3Y(dfObj):
    if (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] <= 36):
        return '1'
    elif (dfObj['Mortality'] == 0):
        return '0'
    elif (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] > 36):
        return '0'
    
def Mortality2Y(dfObj):
    if (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] <= 24):
        return '1'
    elif (dfObj['Mortality'] == 0):
        return '0'
    elif (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] > 24):
        return '0'
    

def Mortality1Y(dfObj):
    if (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] <= 12):
        return '1'
    elif (dfObj['Mortality'] == 0):
        return '0'
    elif (dfObj['Mortality'] == 1) and (dfObj['MtillDeath_censored'] > 12):
        return '0'

dfObj['Mortality5Y'] = dfObj.apply(Mortality5Y, axis = 1)
dfObj['Mortality4Y'] = dfObj.apply(Mortality4Y, axis = 1)   
dfObj['Mortality3Y'] = dfObj.apply(Mortality3Y, axis = 1)   
dfObj['Mortality2Y'] = dfObj.apply(Mortality2Y, axis = 1)   
dfObj['Mortality1Y'] = dfObj.apply(Mortality1Y, axis = 1)   
    
def ICD_nonbenefit(dfObj):
    if (dfObj['Mortality'] == 1) and (dfObj['AppropriateTherapy'] == 0):
        return '1'
    elif (dfObj['Mortality'] == 0):
        return '0'
    elif (dfObj['Mortality'] == 1) and (dfObj['AppropriateTherapy'] == 1):
        return '0'

def ICD_nonbenefit5Y(dfObj):
    if (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] <= 60):
        return '1'
    elif (dfObj['ICDnonbenefit'] == '0'):
        return '0'
    elif (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] > 60):
        return '0'
    
def ICD_nonbenefit4Y(dfObj):
    if (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] <=48):
        return '1'
    elif (dfObj['ICDnonbenefit'] == '0'):
        return '0'
    elif (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] > 36):
        return '0'
    
def ICD_nonbenefit3Y(dfObj):
    if (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] <=36):
        return '1'
    elif (dfObj['ICDnonbenefit'] == '0'):
        return '0'
    elif (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] > 36):
        return '0'

    
def ICD_nonbenefit2Y(dfObj):
    if (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] <=24):
        return '1'
    elif (dfObj['ICDnonbenefit'] == '0'):
        return '0'
    elif (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] > 24):
        return '0'
    
        
def ICD_nonbenefit1Y(dfObj):
    if (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] <=12):
        return '1'
    elif (dfObj['ICDnonbenefit'] == '0'):
        return '0'
    elif (dfObj['ICDnonbenefit'] == '1') and (dfObj['MtillDeath_censored'] > 12):
        return '0'

dfObj['ICDnonbenefit'] = dfObj.apply(ICD_nonbenefit, axis = 1)
dfObj['ICDnonbenefit_5Y'] = dfObj.apply(ICD_nonbenefit5Y, axis = 1)
dfObj['ICDnonbenefit_4Y'] = dfObj.apply(ICD_nonbenefit4Y, axis = 1)
dfObj['ICDnonbenefit_3Y'] = dfObj.apply(ICD_nonbenefit3Y, axis = 1)
dfObj['ICDnonbenefit_2Y'] = dfObj.apply(ICD_nonbenefit2Y, axis = 1)
dfObj['ICDnonbenefit_1Y'] = dfObj.apply(ICD_nonbenefit1Y, axis = 1)

def FU_1Y(dfObj):
    if (dfObj['MtillDeath_censored'] >=12):
        return '1'
    elif (dfObj['MtillDeath_censored'] <12):
        return '0'
def FU_2Y(dfObj):
    if (dfObj['MtillDeath_censored'] >=24):
        return '1'
    elif (dfObj['MtillDeath_censored'] <24):
        return '0'
def FU_3Y(dfObj):
    if (dfObj['MtillDeath_censored'] >=36):
        return '1'
    elif (dfObj['MtillDeath_censored'] <36):
        return '0'
def FU_4Y(dfObj):
    if (dfObj['MtillDeath_censored'] >=48):
        return '1'
    elif (dfObj['MtillDeath_censored'] <48):
        return '0'
def FU_5Y(dfObj):
    if (dfObj['MtillDeath_censored'] >=60):
        return '1'
    elif (dfObj['MtillDeath_censored'] <60):
        return '0'

dfObj['FU_1Y'] = dfObj.apply(FU_1Y, axis = 1)
dfObj['FU_2Y'] = dfObj.apply(FU_2Y, axis = 1)
dfObj['FU_3Y'] = dfObj.apply(FU_3Y, axis = 1)
dfObj['FU_4Y'] = dfObj.apply(FU_4Y, axis = 1)
dfObj['FU_5Y'] = dfObj.apply(FU_5Y, axis = 1)

print('Follow-up of 1 year\n', dfObj['FU_1Y'].value_counts())
print('Follow-up of 5 years\n', dfObj['FU_5Y'].value_counts())

#Select patient with LVEF < 35% and primary prevention ICD
data_sel = dfObj.loc[(dfObj['LowLVEF'] == 1 )| (dfObj['ReducedLVEF'] == 1)]
data = data_sel.loc[(data_sel['Implantation_indication'] == 1)]

#Normalize the continious variables
var = ['Age','BMI', 'Sodium', 'Potassium', 'Kreatinine', 'QRS_Duration']
dfObj=data

X_unnormalised = dfObj[var] 
features_mean = X_unnormalised.mean()
features_std = X_unnormalised.std()
X_norm =((X_unnormalised-features_mean)/features_std)

dfObj['Age_normalised'] = X_norm['Age']
dfObj['BMI_normalised'] = X_norm['BMI']
dfObj['Sodium_normalised'] = X_norm['Sodium']
dfObj['Potassium_normalised'] = X_norm['Potassium']
dfObj['Kreatinine_normalised'] = X_norm['Kreatinine']
dfObj['QRS_Duration_normalised'] = X_norm['QRS_Duration']

dfObj.drop(['Kreatinine', 'Potassium', 'Sodium', 'BMI', 'Age', 'ImplantationDate', 'QRS_Duration'], axis=1, inplace=True)

data = dfObj
#Sum of missing values
pd.set_option('display.max_rows', 100)
print(data.isnull().sum())
print(data.shape)
```
